In this notebook, I compare the VSH parameters based on the bright and faint sample.

```python
import numpy as np
from matplotlib import pyplot as plt
from matplotlib.ticker import MultipleLocator
from my_progs.catalog.vec_mod import vec_mod_calc
%matplotlib inline
%config InlineBackend.figure_format = "svg"
```

```python
minorLocator = MultipleLocator(10)
majorLocator = MultipleLocator(50)
name01 = ["$G_1$", "$G_2$", "$G_3$", "$R_1$", "$R_2$", "$R_3$"]
name02 = ["$E_{22}^R$", "$E_{22}^I$", "$E_{21}^R$", "$E_{21}^I$", "$E_{20}$",
         "$M_{22}^R$", "$M_{22}^I$", "$M_{21}^R$", "$M_{21}^I$", "$M_{20}$"]


barwidth = 0.15
loc = 0.15

pos1 = np.arange(len(name01)) - 1 * loc
bgt_pos1 = np.arange(len(name01))
fnt_pos1 = np.arange(len(name01)) + 1 * loc

pos2 = np.arange(len(name02)) - 1 * loc
bgt_pos2 = np.arange(len(name02))
fnt_pos2 = np.arange(len(name02)) + 1 * loc
```

For ICRF1

```python
pmt1, err1 = np.genfromtxt("../logs/icrf1_gaiadr2_vsh02.log",
                           usecols=(1, 2), skip_header=1, unpack=True)
pmt1bgt, err1bgt = np.genfromtxt("../logs/icrf1_gaiadr2_bgt_vsh02.log",
                                 usecols=(1, 2), skip_header=1, unpack=True)
pmt1fnt, err1fnt = np.genfromtxt("../logs/icrf1_gaiadr2_fnt_vsh02.log",
                                 usecols=(1, 2), skip_header=1, unpack=True)
```

```python
# Rotation and Glide
fig, ax = plt.subplots()
ax.bar(pos1, 2 * err1[:6], bottom=pmt1[:6]-err1[:6], width=barwidth,
       color="g", ecolor="black", label="All")
ax.bar(bgt_pos1, 2 * err1bgt[:6], bottom=pmt1bgt[:6]-err1bgt[:6], width=barwidth,
       color="b", ecolor="black", label="G < 18.7")
ax.bar(fnt_pos1, 2 * err1fnt[:6], bottom=pmt1fnt[:6]-err1fnt[:6], width=barwidth,
       color="r", ecolor="black", label="G >= 18.7")

ax.set_xticks(bgt_pos1)
ax.set_xticklabels(name01, fontsize=15)
ax.set_xlim([-1, len(pos1)])
ax.set_ylim([-100, 300])
ax.set_ylabel("Rotation and Glide ($\\mathrm{\mu as}$)")
ax.xaxis.set_ticks_position("both")
ax.yaxis.set_ticks_position("both")
ax.yaxis.grid()  # horizontal lines
ax.yaxis.set_minor_locator(minorLocator)
ax.set_title("ICRF1 $-$ Gaia DR2")
ax.legend(fontsize="x-small")
```

```python
# Quadrupole
fig, ax = plt.subplots()
ax.bar(pos2, 2 * err1[6:], bottom=pmt1[6:]-err1[6:], width=barwidth,
       color="g", ecolor="black", label="All")
ax.bar(bgt_pos2, 2 * err1bgt[6:], bottom=pmt1bgt[6:]-err1bgt[6:], width=barwidth,
       color="b", ecolor="black", label="G < 18.7")
ax.bar(fnt_pos2, 2 * err1fnt[6:], bottom=pmt1fnt[6:]-err1fnt[6:], width=barwidth,
       color="r", ecolor="black", label="G >= 18.7")

ax.set_xticks(bgt_pos2)
ax.set_xticklabels(name02, fontsize=15)
ax.set_xlim([-1, len(pos2)])
ax.set_ylim([-200, 250])
ax.set_ylabel("Quadrupole ($\\mathrm{\mu as}$)")
ax.xaxis.set_ticks_position("both")
ax.yaxis.set_ticks_position("both")
ax.yaxis.grid()  # horizontal lines
ax.yaxis.set_minor_locator(minorLocator)
ax.set_title("ICRF1 $-$ Gaia DR2")
ax.legend(fontsize="x-small")
```
