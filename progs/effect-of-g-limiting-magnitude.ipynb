```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = "svg"
```

The number of common sources as a function of limiting magnitude.

```python
gmag1, N0_1, N1_1, X0_1, rate1 = np.genfromtxt(
    "../logs/g-limiting-mag-stat-icrf1.log", unpack=True)
gmag2, N0_2, N1_2, X0_2, rate2 = np.genfromtxt(
    "../logs/g-limiting-mag-stat-icrf2.log", unpack=True)
gmag3sx, N0_3sx, N1_3sx, X0_3sx, rate3sx = np.genfromtxt(
    "../logs/g-limiting-mag-stat-icrf3sx.log", unpack=True)
gmag3k, N0_3k, N1_3k, X0_3k, rate3k = np.genfromtxt(
    "../logs/g-limiting-mag-stat-icrf3k.log", unpack=True)
gmag3xka, N0_3xka, N1_3xka, X0_3xka, rate3xka = np.genfromtxt(
    "../logs/g-limiting-mag-stat-icrf3xka.log", unpack=True)
```

```python
fig, ax = plt.subplots()

ax.plot(gmag1, N0_1, "g-v", label="ICRF1")
ax.plot(gmag2, N0_2, "m-s", label="ICRF2")
ax.plot(gmag3sx, N0_3sx, "r-x", label="ICRF3 SX")
ax.plot(gmag3k, N0_3k, "y-o", label="ICRF3 K")
ax.plot(gmag3xka, N0_3xka, "b-*", label="ICRF3 XKa")

ax.axis([17.4, 21.1, 0, 3000])
ax.set_xlabel("$G$ (mag)")
ax.set_ylabel("Number of common sources")

# ax.set_yscale("log")

ax.xaxis.set_ticks_position("both")
ax.yaxis.set_ticks_position("both")

ax.grid()
ax.legend()
```

```python
fig, ax = plt.subplots()

ax.plot(gmag1, N1_1, "g-v", label="ICRF1")
ax.plot(gmag2, N1_2, "m-s", label="ICRF2")
ax.plot(gmag3sx, N1_3sx, "r-x", label="ICRF3 SX")
ax.plot(gmag3k, N1_3k, "y-o", label="ICRF3 K")
ax.plot(gmag3xka, N1_3xka, "b-*", label="ICRF3 XKa")

ax.axis([17.4, 21.1, 0, 2500])
ax.set_xlabel("$G$ (mag)")
ax.set_ylabel("Number of used sources")

# ax.set_yscale("log")

ax.xaxis.set_ticks_position("both")
ax.yaxis.set_ticks_position("both")

ax.grid()
ax.legend(fontsize="x-small")
```

```python
fig, ax = plt.subplots()

ax.plot(gmag1, N0_1 * rate1, "g-v", label="ICRF1")
ax.plot(gmag2, N1_2 * rate2, "m-s", label="ICRF2")
ax.plot(gmag3sx, N1_3sx * rate3sx, "r-x", label="ICRF3 SX")
ax.plot(gmag3k, N1_3k * rate3k, "y-o", label="ICRF3 K")
ax.plot(gmag3xka, N1_3xka * rate3xka, "b-*", label="ICRF3 XKa")

ax.axis([17.4, 21.1, 0, 400])
ax.set_xlabel("$G$ (mag)")
ax.set_ylabel("NO. outliers")

ax.xaxis.set_ticks_position("both")
ax.yaxis.set_ticks_position("both")

ax.grid()
ax.legend(fontsize="x-small")
```
